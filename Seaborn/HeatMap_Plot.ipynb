```python
var = np.linspace(1,10,10).reshape(2,5)
var
```

```python
arr = np.array([['a0','a1', 'a2','a3','a4'],
                ['b0','b1','b2', 'b3','b4']])
arr
```

```python
sns.heatmap(var, vmin = 0, vmax =14, cmap ='gist_heat',annot =arr,fmt = 's')
plt.show()
```

```python
y = {'fontsize':10, 'color':'orange'}

sns.heatmap(var, vmin = 0, vmax =14, cmap ='gist_heat',annot =arr,fmt = 's',
            annot_kws = y)
plt.show()
```

```python
y = {'fontsize':10, 'color':'orange'}

sns.heatmap(var, vmin = 0, vmax =14, cmap ='gist_heat',annot =arr,fmt = 's',
            annot_kws = y, linewidth = 4)
plt.show()
```

```python
y = {'fontsize':10, 'color':'orange'}

sns.heatmap(var, vmin = 0, vmax =14, cmap ='gist_heat',annot =arr,fmt = 's',
            annot_kws = y, linewidth = 4,linecolor = 'y')
plt.show()
```

```python
y = {'fontsize':10, 'color':'orange'}

sns.heatmap(var, vmin = 0, vmax =14, cmap ='gist_heat',annot =arr,fmt = 's',
            annot_kws = y, linewidth = 4,linecolor = 'b')
plt.show()
```

```python
y = {'fontsize':10, 'color':'orange'}

sns.heatmap(var, vmin = 0, vmax =14, cmap ='gist_heat',annot =arr,fmt = 's',
            annot_kws = y, linewidth = 4,linecolor = 'b', cbar = False)
plt.show()
```

```python
y = {'fontsize':10, 'color':'orange'}

sns.heatmap(var, vmin = 0, vmax =14, cmap ='gist_heat',annot =arr,fmt = 's',
            annot_kws = y, linewidth = 4,linecolor = 'b', cbar = False, xticklabels = False)
plt.show()
```

```python
y = {'fontsize':10, 'color':'orange'}

sns.heatmap(var, vmin = 0, vmax =14, cmap ='gist_heat',annot =arr,fmt = 's',
            annot_kws = y, linewidth = 4,linecolor = 'b', cbar = False, yticklabels = False)
plt.show()
```

```python
y = {'fontsize':10, 'color':'orange'}

v = sns.heatmap(var, vmin = 0, vmax =14, cmap ='gist_heat',annot =arr,fmt = 's',
            annot_kws = y, linewidth = 4,linecolor = 'b', cbar = False, xticklabels = False)
v.set(xlabel='Python', ylabel ='Program')
plt.show()
```
